```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import dlti, freqz

# Define your signal x(n)
n = np.arange(0, 10)
x_n = np.sin(0.1 * np.pi * n)

# Compute the z-transform using dlti
system = dlti(x_n, [1.0], dt=1)
w, X_z = freqz(x_n, worN=8000)

# Plot x(n)
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.stem(n, x_n, use_line_collection=True)
plt.title('Signal x(n)')
plt.xlabel('n')
plt.ylabel('x(n)')

# Plot |X(z)|
plt.subplot(1, 2, 2)
plt.plot(w, abs(X_z))
plt.title('Magnitude of X(z)')
plt.xlabel('Frequency (rad/sample)')
plt.ylabel('|X(z)|')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import dlti, freqz

# Define a simple sequence
n = np.arange(0, 5)
x = np.array([1, 2, 3, 4, 5])

# Time shifted sequence
k = 1
x_shifted = np.roll(x, k)
x_shifted[:k] = 0

# Time reversed sequence
x_reversed = x[::-1]

# Compute the z-transform of original sequence
sys_x = dlti(x, [1])  # [1] is the denominator for z-transform
w_x, H_x = freqz(x, worN=8000)

# Compute the z-transform of time-shifted sequence
sys_shifted = dlti(x_shifted, [1])
w_shifted, H_shifted = freqz(x_shifted, worN=8000)

# Compute the z-transform of time-reversed sequence
sys_reversed = dlti(x_reversed, [1])
w_reversed, H_reversed = freqz(x_reversed, worN=8000)

# Plotting
plt.figure(figsize=(15, 5))

# Original and Time-Shifted
plt.subplot(1, 2, 1)
plt.plot(w_x, np.abs(H_x), label="Original Signal")
plt.plot(w_shifted, np.abs(H_shifted), label=f"Time Shifted Signal (by {k} units)", linestyle="--")
plt.title("Original vs. Time Shifted Signal in z-domain")
plt.xlabel("Frequency (rad/sample)")
plt.ylabel("Magnitude")
plt.legend()

# Original and Time-Reversed
plt.subplot(1, 2, 2)
plt.plot(w_x, np.abs(H_x), label="Original Signal")
plt.plot(w_reversed, np.abs(H_reversed), label="Time Reversed Signal", linestyle="--")
plt.title("Original vs. Time Reversed Signal in z-domain")
plt.xlabel("Frequency (rad/sample)")
plt.ylabel("Magnitude")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import dlti, freqz

# Define a simple sequence
n = np.arange(0, 5)
x = np.array([1, 2, 3, 4, 5])

# Scaled sequence
a = 0.5
x_scaled = a**n * x

# Compute the z-transform of original sequence
sys_x = dlti(x, [1])
w_x, H_x = freqz(x, worN=8000)

# Compute the z-transform of scaled sequence
sys_scaled = dlti(x_scaled, [1])
w_scaled, H_scaled = freqz(x_scaled, worN=8000)

# Plotting
plt.figure(figsize=(15, 5))

# Original and Scaled in time domain
plt.subplot(1, 2, 1)
plt.stem(n, x, basefmt=" ", use_line_collection=True, label="Original Signal")
plt.stem(n, x_scaled, basefmt=" ", use_line_collection=True, linefmt='r--', markerfmt='ro', label=f"Scaled Signal (a = {a})")
plt.title("Original vs. Scaled Signal in Time Domain")
plt.xlabel("n")
plt.ylabel("Magnitude")
plt.legend()

# Original and Scaled in z-domain
plt.subplot(1, 2, 2)
plt.plot(w_x, np.abs(H_x), label="Original Signal")
plt.plot(a*w_scaled, np.abs(H_scaled), label=f"Scaled Signal in z-domain (a = {a})", linestyle="--")
plt.title("Original vs. Scaled Signal in z-domain")
plt.xlabel("Frequency (rad/sample)")
plt.ylabel("Magnitude")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a simple sequence
n = np.arange(0, 5)
x = np.array([1, 2, 3, 4, 5])

# Define a mesh of z values (complex plane)
re = np.linspace(-2, 2, 400)
im = np.linspace(-2, 2, 400)
Re, Im = np.meshgrid(re, im)
Z = Re + 1j*Im

# Compute the z-transform
X_z = np.zeros_like(Z)
for idx, val in enumerate(x):
    X_z += val * (Z**(-idx))

# Extract the magnitude
magnitude = np.abs(X_z)

# Plotting in 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(Re, Im, magnitude, cmap='viridis')

# Setting labels
ax.set_xlabel('Re(z)')
ax.set_ylabel('Im(z)')
ax.set_zlabel('|X(z)|')
ax.set_title('3D Visualization of Magnitude of Z-transform')

plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a discrete-time sinusoidal sequence
n = np.arange(0, 100)
f = 0.05  # frequency of the sine wave
x = np.sin(2 * np.pi * f * n)

# Define a mesh of z values (complex plane)
re = np.linspace(-2, 2, 400)
im = np.linspace(-2, 2, 400)
Re, Im = np.meshgrid(re, im)
Z = Re + 1j*Im

# Compute the z-transform
X_z = np.zeros_like(Z)
for idx, val in enumerate(x):
    X_z += val * (Z**(-idx))

# Extract the magnitude
magnitude = np.abs(X_z)

# Plotting in 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(Re, Im, magnitude, cmap='viridis', edgecolor='none')

# Setting labels
ax.set_xlabel('Re(z)')
ax.set_ylabel('Im(z)')
ax.set_zlabel('|X(z)|')
ax.set_title('3D Visualization of Magnitude of Z-transform of a Sinusoid')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate a finite white Gaussian noise sequence
N = 100
x = np.random.normal(0, 1, N)

# Z-transform approximation for discrete sequence
def Z_transform(x, z):
    return sum([x[n] * z**(-n) for n in range(len(x))])

# Values for z = re^{j\theta}, r from 0 to 2 and theta from 0 to 2*pi
radius = np.linspace(0.05, 2, 100)
theta = np.linspace(0, 2*np.pi, 100)

R, Theta = np.meshgrid(radius, theta)
Z = R * np.exp(1j * Theta)
X_z = np.array([[Z_transform(x, z) for z in row] for row in Z])

# Plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(R * np.cos(Theta), R * np.sin(Theta), np.abs(X_z), cmap='viridis')

ax.set_title('Magnitude of Approximated Z-Transform of White Gaussian Noise')
ax.set_xlabel('Real')
ax.set_ylabel('Imaginary')
ax.set_zlabel('|X(z)|')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Values for z = re^{j\theta}, r from 0.5 to 2 and theta from 0 to 2*pi
radius = np.linspace(0.5, 2, 100)
theta = np.linspace(0, 2*np.pi, 100)

R, Theta = np.meshgrid(radius, theta)
Z = R * np.exp(1j * Theta)

# The z-transform of the delta function is 1 everywhere
X_z = np.ones_like(Z)

# Plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(R * np.cos(Theta), R * np.sin(Theta), np.abs(X_z), cmap='viridis')

ax.set_title('Magnitude of Z-Transform of Delta Function')
ax.set_xlabel('Real')
ax.set_ylabel('Imaginary')
ax.set_zlabel('|X(z)|')
plt.show()
```
